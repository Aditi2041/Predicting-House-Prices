```python
#Checking the correlation of the AgeSold with SalePrice which is negative implying that old houses have low SalePrice
house_df['AgeSold'].corr(house_df['SalePrice'])
```

```python
# Missing Values 
# Checking those variables with missing values greater than zero
house_df_na = house_df.isnull().sum()
house_df_na = house_df_na[house_df_na>0]
house_df_na.sort_values(ascending=False)
```

```python
# Missing values imputed with None
house_df['PoolQC'] = house_df['PoolQC'].fillna('None')
```

```python
# Missing values imputed with None
house_df['MiscFeature'] = house_df['MiscFeature'].fillna('None')
```

```python
# Missing values imputed with None
house_df['Alley'] = house_df['Alley'].fillna('None')
```

```python
# Missing values imputed with None
house_df['Fence'] = house_df['Fence'].fillna('None')
```

```python
# Missing values imputed with None
house_df['FireplaceQu'] = house_df['FireplaceQu'].fillna('None')
```

```python
# Since the area of each street connected to the house most likely have a similar area to other houses in its
# neighborhood , we can fill in missing values by the median LotFrontage of the neighborhood.
house_df['LotFrontage'] = house_df.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
```

```python
# Missing values imputed with None
for col in ('GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'):
    house_df[col] = house_df[col].fillna('None')
```

```python
# Missing values imputed with 0 as it is a numeric variable
house_df['GarageYrBlt'] = house_df['GarageYrBlt'].fillna(0)
```

```python
# Missing values imputed with None
for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    house_df[col] = house_df[col].fillna('None')
```

```python
# For this categorical feature , Utilities ,  all records are "AllPub", except for one and hence it won't help in predictive
# modelling , thus dropping it
house_df = house_df.drop(['Utilities'], axis=1)
# MasVnrArea and MasVnrType are strongly correlated with YearBuilt and OverallQual , thus removing it
house_df = house_df.drop(['MasVnrArea'], axis=1)
house_df = house_df.drop(['MasVnrType'], axis=1)
```

```python
# It has only 1 NA value , hence imputing the mode
house_df['Electrical'] = house_df['Electrical'].fillna(house_df['Electrical'].mode()[0])
```

```python
#checking if there are anymore null values
print(house_df.isnull().sum().max())
```

```python
house_df.shape
```

```python
# inspecting the target variable
from scipy.stats import norm
sns.distplot(house_df['SalePrice'], fit=norm)
```

```python
# The SalePrice is slightly skewed and hence performing log transformation
house_df['SalePrice'] = np.log(house_df['SalePrice'])
```

```python
# Inspecting once again
sns.distplot(house_df['SalePrice'], fit=norm)
```

```python
# Differentiating the variables into numerical and categorical
cat_features = house_df.select_dtypes(include = ["object"]).columns
num_features = house_df.select_dtypes(exclude = ["object"]).columns
house_df_num = house_df[num_features]
house_df_cat = house_df[cat_features]
```

```python
# checking for skew in the numerical variables
from scipy.stats import skew 
skewness = house_df_num.apply(lambda x: skew(x))
skewness = skewness[abs(skewness)>0.5]
skew_features = house_df[skewness.index]
skew_features.columns
```

```python
# correcting those features by log transformation
skew_features = np.log(skew_features)
```

```python
# Creating dummies for the categorical variables
house_df_cat = pd.get_dummies(house_df_cat , drop_first = True)
```

```python
# Concatenating the nmerical variables that has been transformed and the dummy variables
total_features = pd.concat([house_df_cat,house_df_num],axis=1)
total_features.shape
```

```python
# Assigning SalePrice as the target variable
y = total_features['SalePrice']
```

```python
# Dropping SalePrice from the predictors
x = total_features.drop(['SalePrice'] , axis = 1)
```

```python
# Splitting the data in test and train datasets. Putting random_state = 100, so that it will always generate same splits.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
#importing the libraries for regression
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
import os
```

```python
# lasso regression
from sklearn import metrics
lm = Lasso(alpha=0.001)
lm.fit(X_train, y_train)

# predict
y_train_pred = lm.predict(X_train)
print(metrics.r2_score(y_true=y_train, y_pred=y_train_pred))
y_test_pred = lm.predict(X_test)
print(metrics.r2_score(y_true=y_test, y_pred=y_test_pred))
```

```python
# lasso model parameters
model_parameters = list(lm.coef_)
model_parameters.insert(0, lm.intercept_)
model_parameters = [round(x, 3) for x in model_parameters]
cols = x.columns
cols = cols.insert(0, "constant")
list(zip(cols, model_parameters))
```

```python
# grid search CV

# set up cross validation scheme
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
folds = KFold(n_splits = 5, shuffle = True, random_state = 4)

def rmse_cv(model):
    rmse= np.sqrt(-cross_val_score(model, X_train, y_train, scoring="neg_mean_squared_error", cv = 5))
    return(rmse)

# specify range of hyperparameters
params = {'alpha': [0.001, 0.01, 1.0, 5.0, 10.0]}

# grid search
# lasso model
model = Lasso()
model_cv_lasso = GridSearchCV(estimator = model, param_grid = params, 
                        scoring= 'r2', 
                        cv = folds, 
                        return_train_score=True, verbose = 1)            
model_cv_lasso.fit(X_train, y_train) 
```

```python
# Calculating root mean square error of Lasso Regression
rmse_cv(model_cv_lasso).mean()
```

```python
cv_results = pd.DataFrame(model_cv_lasso.cv_results_)
cv_results.head()
```

```python
# plotting the test and train
cv_results['param_alpha'] = cv_results['param_alpha'].astype('float32')
plt.plot(cv_results['param_alpha'], cv_results['mean_train_score'])
plt.plot(cv_results['param_alpha'], cv_results['mean_test_score'])
plt.xlabel('alpha')
plt.ylabel('r2 score')
plt.legend(['train score', 'test score'], loc='upper left')
plt.show()
```
